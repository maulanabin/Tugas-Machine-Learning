```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Machine Learning/Pertemuan 6/mushrooms.csv')
df.head()
```

```python
df = df.astype('category')
df.dtypes
```

```python
# Menggunakan LabelEncoder untuk mengonversi nilai kategori menjadi angka
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for column in df.columns:
    df[column] = labelencoder.fit_transform(df[column])

df.head()
```

Dari tabel di atas dapat dilihat bahwa kolom 'veil-type' hanya memiliki satu nilai unik sehingga tidak akan memberikan kontribusi apa pun pada data. Jadi bisa langsung dihapus saja.

```python
df = df.drop(["veil-type"],axis=1)
```

Di sini 'axis-1' dapat diartikan sebagai menghapus seluruh kolom (yaitu secara vertikal), jika itu adalah axis-0 akan menjatuhkan seluruh baris (yaitu secara horizontal).

```python
from sklearn.model_selection import train_test_split
# kolom "class" sebagai numpy array.
y = df["class"].values
# Semua data kecuali kolom "class".
x = df.drop(["class"], axis=1).values
# Memisahkan data untuk train dan test.
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

**Classification Methods**

1. Logistic Regression Classification

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver="liblinear")
lr.fit(x_train,y_train)

print("Test Accuracy: {}%".format(round(lr.score(x_test,y_test)*100,2)))
```

2. Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)

print("Test Accuracy: {}%".format(round(dt.score(x_test,y_test)*100,2)))
```

```python
from sklearn.tree import DecisionTreeClassifier

# Instansiasi dt_entropy dan dt_gini serta mengatur 'entropy' dan 'gini' sebagai kriteria informasi
dt_entropy = DecisionTreeClassifier(max_depth=8, criterion='entropy', random_state=1)
dt_gini = DecisionTreeClassifier(max_depth=8, criterion='gini', random_state=1)

# Fit dt_entropy dan dt_gini ke training set
dt_entropy.fit(x_train, y_train)
dt_gini.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
 
# Menggunakan dt_entropy dan dt_gini untuk memprediksi label set pengujian
y_pred= dt_entropy.predict(x_test)
y_pred_gini = dt_gini.predict(x_test)

# Mengevaluasi accuracy_entropy dan accuracy_gini
accuracy_entropy = accuracy_score(y_test, y_pred)
accuracy_gini = accuracy_score(y_test, y_pred_gini)

# Cetak accuracy_entropy dan accuracy_gini
print('Akurasi dicapai dengan menggunakan entropi: ', accuracy_entropy)
print('Akurasi dicapai dengan menggunakan gini index: ', accuracy_gini)
```
