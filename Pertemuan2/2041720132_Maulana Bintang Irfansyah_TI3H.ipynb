This will show:

<ul>
    <li>the count of that variable</li>

    <li>IQR (Interquartile Range: 25%, 50% dan 75%)</li>

 Kita dapat meutilisasi method "describe" seperti berikut ini:


```python
df.describe()
```

Secara default describe melakukan ignore terhada variabel bertype objek. Kita dapat menggunakan method describe pada variable object dengan cara sebagai berikut


```python
df.describe(include=['object'])
```


<h3>Value Counts</h3>
<p>Penghitungan nilai adalah cara yang baik untuk memahami berapa banyak unit dari setiap karakteristik/variabel yang kita miliki. Kita bisa menerapkan metode "value_counts" pada kolom 'drive-wheels'. Jangan lupa metode "value_counts" hanya berfungsi pada seri Pandas, bukan Pandas Dataframes. Sehingga, kita hanya menyertakan satu tanda kurung "df['petal_length']" bukan dua tanda kurung "df[['petal_length']]".</p> 


```python
df['sepal_length'].value_counts()
```

Kita dapat melakukan konversi kedalam dataframe dengan cara sebagai berikut :


```python
df['petal_length'].value_counts().to_frame()
```

Mari ulangi langkah di atas tetapi simpan hasilnya ke dataframe "sepal_length_counts" dan ganti nama kolom 'sepal_length' menjadi 'value_counts'. 


```python
sepal_length_counts = df['sepal_length'].value_counts().to_frame()
sepal_length_counts.rename(columns={'sepal_length': 'value_counts'}, inplace=True)
sepal_length_counts
```

Sekarang mari kita ganti nama indeks menjadi 'sepal_length': 


```python
sepal_length_counts.index.name = 'sepal_length'
sepal_length_counts
```

Kita dapat mengulangi proses di atas untuk variabel 'petal_length'. 


```python
# engine-location as variable
engine_loc_counts = df['petal_length'].value_counts().to_frame()
engine_loc_counts.rename(columns={'petal_length': 'value_counts'}, inplace=True)
engine_loc_counts.index.name = 'petal_length'
engine_loc_counts.head(10)
```
